```python
import pandas as pd
import numpy as np
import matplotlib 
```

```python
jokeset = pd.read_csv('Desktop/emmys/emmy_jokes.csv')
jokeset
```

```python
running_times = pd.read_csv('Desktop/emmys/emmy_running_times.csv')
running_times = running_times.set_index('show')
running_times
```

```python
#calculating the number of jokes per show
blackish_len = jokeset.get('Blackish (S07)').value_counts().sum()
cobra_len = jokeset.get('Cobra Kai (S03)').value_counts().sum()
emily_len = jokeset.get('Emily in Paris (S01)').value_counts().sum()
hacks_len = jokeset.get('Hacks (S01)').value_counts().sum()
pen_len = jokeset.get('Pen15 (S02)').value_counts().sum()
lasso_len = jokeset.get('Ted Lasso (S01)').value_counts().sum()
flight_len = jokeset.get('The Flight Attendant (S01)').value_counts().sum()
kominksy_len = jokeset.get('The Kominsky Method (S02)').value_counts().sum()
num_jokes = np.array(list([blackish_len,cobra_len,emily_len,hacks_len,pen_len,lasso_len,flight_len,kominksy_len]))
num_jokes
```

```python
#adding a new column, "num_jokes" to the "running times" dataset to create a new dataset called "jokes_per_min"
jokes_per_min = running_times
jokes_per_min['num_jokes'] = num_jokes
jokes_per_min['jokes_per_minute'] = num_jokes/jokes_per_min.get('running_time_minutes')
jokes_per_min
```

```python
#finding the show with the highest number of jokes per minute
max_jokes = jokes_per_min.get('jokes_per_minute').max()
jokes_per_min[jokes_per_min['jokes_per_minute'] == max_jokes].reset_index().get('show')[0]
```
